```python
import pandas as pd
path = r'./train_test_nq.jsonl'

inputs = pd.read_json(path, orient='records', lines=True)
train_10 = inputs.query("split == 'train'").sample(10, random_state=42)
test_10  = inputs.query("split == 'test'").sample(10, random_state=42)
inputs = pd.concat([train_10, test_10], ignore_index=True)
```

```python
res = r'./train_test_nq_split.jsonl'
inputs.to_json(res, orient='records', lines=True)
```
